```python
import h5py
```

```python
from transformers import T5Tokenizer, T5EncoderModel
import torch
import re
```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
model = T5EncoderModel.from_pretrained('Rostlab/prot_t5_xl_half_uniref50-enc').to(device)
tokenizer = T5Tokenizer.from_pretrained('Rostlab/prot_t5_xl_half_uniref50-enc')
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/PBL/datasets/exp_proteins_clustered_70_filter_lens_25-75_undersampled.tsv', sep='\t')
seqs = [s for s in df['Sequence']]
seqs_spaces = [" ".join(list(re.sub(r"[UZOB]", "X", sequence))) for sequence in seqs]
lens = [len(s) for s in seqs]
ids = tokenizer(seqs_spaces, add_special_tokens=True, padding="longest")
```

```python
input_ids= torch.tensor(ids['input_ids']).to(device)
attention_mask= torch.tensor(ids['attention_mask']).to(device)
```

```python
batch_size = 24
model.eval()
with torch.no_grad():
  with h5py.File('/content/drive/MyDrive/PBL/embeddings_exp_proteins_clustered_70_filter_lens_25-75_undersampled_1.h5', 'w') as f:
    for i in range(0, len(seqs), batch_size):
      batch_ids = input_ids[i:i+batch_size].to('cuda')
      batch_att = attention_mask[i:i+batch_size].to('cuda')
      batch_output = model(input_ids=batch_ids, attention_mask=batch_att)
      pp_embeddings = per_protein_embeddings(batch_output, i)
      f.create_dataset(f'Batch {i // batch_size}', data=pp_embeddings)
      print('Done with Batch ' + str(i // batch_size))
      f.flush()
```

```python
import numpy as np

def per_protein_embeddings(batch_result, i):
  batch_hidden_states = batch_result.last_hidden_state
  per_residue_embeddings = list()

  for j in range(0, len(batch_hidden_states)):
    curr_len = lens[i+j]
    per_residue_embeddings.append(batch_hidden_states[j][:curr_len])

  per_protein_embeddings = [embedding.mean(dim=0).cpu() for embedding in per_residue_embeddings]
  return np.array(per_protein_embeddings)
```
